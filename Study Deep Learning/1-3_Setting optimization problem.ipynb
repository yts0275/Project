#### 신경망 훈련 시, training 속도를 높일 수 있는 방법 중 하나 = 입력 값을 normalize 시키는 것

입력 값을 표준화 시키는 절차는 2가지 단계를 따른다.

1. 평균값을 빼거나 0으로 만드는 단계
2. 변동 편차를 표준화 하는 단계

#### 입력 값의 특성을 표준화 하려는 이유

- 표준화되지 않은 입력 특성을 이용하면, 비용함수가 푹 퍼진 그릇 모양이 될 가능성이 크다. ( 그림 참조 )
    - x1 = 0 ~ 1000, x2 = 0 ~ 1 의 범위를 지닐 때
    - 각각의 w1, w2 값의 범위가 매우 다를 것, 이 때 접시모양이 나옴
  
 
- 표준화된 입력 특성을 이용하면, 평균적으로 더 대칭 성향을 띄우는 모양이 될 것
  
  
- 접시모양의 비용함수에 경사 하강법 적용 시, 매우 작은 학습률을 지정해야 함
    - 더 많은 단계를 거쳐서 최소로 향하기 때문에  
    
    
- 구형의 비용함수에 경사 하강법 적용 시,
    - 어디서 시작하더라도 바로 최소점에 도달 가능


### Vanishing / Exploding Gradients
신경망을 교육시키는 것의 큰 문제점들
- 데이터가 사라지는 경우
- 기울기가 폭발적으로 증가하는 경우
  
  
심층신경망을 training 하는 경우, 미분값이나 기울기가 굉장히 큰 값을 갖게 되거나, 기하급수적으로 작아질 수도 있다.
